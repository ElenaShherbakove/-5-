```python
import seaborn as sns
```

```python
df= pd.read_csv('sample_data/california_housing_train.csv')
```

```python
peng = sns.load_dataset("penguins")
```

```python
peng.head()
```

```python
sns.scatterplot(data=df, x="longitude", y="latitude")
```

```python
sns.relplot(x="longitude", y="median_house_value", kind="line", data=df)
```

```python
sns.histplot(data=df, x="housing_median_age")
```

```python
sns.scatterplot(data=peng, x="flipper_length_mm", y="body_mass_g", hue="island")
```

```python
sns.scatterplot(data=peng, x="body_mass_g", y="bill_depth_mm", size='island')
```

```python
sns.scatterplot(data=peng, x="bill_length_mm", y="flipper_length_mm", size='sex', hue="species")
```

* Использовать PairGrid с типом графика на ваш выбор

```python
g = sns.PairGrid(peng)
g.map(sns.scatterplot)
```

```python
sns.histplot(data=peng, x='flipper_length_mm')
```

```python
g = sns.PairGrid(peng)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
sns.histplot(data=peng, x='body_mass_g')
```

```python
g = sns.PairGrid(peng, vars=["bill_length_mm", "flipper_length_mm", 'body_mass_g'], hue="species")
g.map_offdiag(sns.scatterplot)
g.map_diag(sns.histplot)
g.add_legend()
```
